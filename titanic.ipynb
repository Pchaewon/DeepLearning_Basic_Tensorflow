##### 회귀선이 있는 산점도
> 두 변수 사이의 산점도를 그리고 선형 회귀 분석에 의해 회귀선을 그림

```python
import seaborn as sns
import matplotlib.pyplot as plt

# grid 스타일 테마 정하기
sns.set_style('darkgrid')

# 그래프 객체 생성
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1, 2, 1)

# 그래프 그리기 - 선형회귀선 
sns.regplot(x='Age',
           y='Fare',
           data=train,
           ax=ax1,
           fit_reg=True)# fig_reg를 True로 하면 회귀선 표시
```

##### 히스토그램/커널 밀도 그래프
> distplot() : 단변수 데이터의 분포 확인시 이용.

```python
# 그래프 객체 생성
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

sns.displot(train['Age'],ax=ax1)

sns.kdeplot(x='Fare',data=train, ax=ax2)
sns.histplot(x='Fare',data=train, ax=ax3)

ax1.set_title('fare-hist+kde')
ax2.set_title('fare-kdeplot')
ax3.set_title('fare-histhist')

plt.show()
```
